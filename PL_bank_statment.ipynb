```python
import pandas as pd
import re
from collections import Counter
```

```python
statment = statment.iloc[:,:-1]
statment.columns = ['date', 'value_date', 'description', 'type', 'amount', 'currency', 'balance']
```

```python
statment.head()
```

```python
statment.date = pd.to_datetime(statment.date)
statment.value_date = pd.to_datetime(statment.value_date)
```

```python
statment = statment.drop(columns=['value_date', 'currency'])
```

```python
statment.head()
```

```python
df = statment.copy()
df.head()
```

```python
df.description = df.description.apply(lambda x: {item.split(':')[0]: item.split(':')[1] for item in x[:-1].split('|') if ':' in item})
#df.description = df.description.apply(lambda x: [item for item in x[:-1].split('|')])
```

```python
new_columns = list(set(df.description.apply(lambda x: list(x.keys())).sum()))
```

```python
for column in new_columns:
    df[column] = df.description.apply(lambda x: x[column] if column in list(x.keys()) else None)
df = df.drop(columns='description')
```

```python
df.to_excel('G:/Мой диск/04 FS/2020 M 07/PL/ACCOUNTS-HISTORY.xlsx',  index=False)
```
