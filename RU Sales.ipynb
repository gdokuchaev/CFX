```python
stores = stores[['iiko_store_id','full_name']]
```

```python
stores = stores.set_index('iiko_store_id')
stores.index.name = None
stores
```

```python
pay = pay[['UniqOrderId.Id','VAT.Sum', 'DiscountSum']]
pay = pay.set_index('UniqOrderId.Id')
pay.index.name = None
pay
```

```python
base = base[['UniqOrderId.Id', 'OpenDate.Typed', 'OrderDeleted', 'fullSum', 'Storned', 'Store.Id']]
```

```python
base = base.loc[(base.Storned == 'FALSE') & (base.OrderDeleted == 'NOT_DELETED')]
base = base[['UniqOrderId.Id', 'OpenDate.Typed', 'fullSum', 'Store.Id']]
base = base.set_index('UniqOrderId.Id')
base.index.name = None
base
```

```python
base = base.join(pay).join(stores, on='Store.Id')
```

```python
base = base.drop(columns=['Store.Id', 'VAT.Sum'])[['full_name', 'OpenDate.Typed', 'fullSum', 'DiscountSum']]
```

```python
base
```

```python
base['OpenDate.Typed'] = pd.to_datetime(base['OpenDate.Typed'], format='%Y-%m-%d')
```

```python
base['year'] = base['OpenDate.Typed'].dt.year
base['month'] = base['OpenDate.Typed'].dt.month
base['day'] = base['OpenDate.Typed'].dt.day
```

```python
base = base[['full_name', 'year', 'month', 'day', 'fullSum', 'DiscountSum']]
base
```

```python
orders = base.groupby(by=['full_name', 'year', 'month'], as_index= False).count()[['fullSum']].rename(columns={'fullSum': 'orders'})
pivot = base.groupby(by=['full_name', 'year', 'month'], as_index= False).sum().join(orders)
```

```python
pivot = pivot[['full_name', 'year', 'month', 'orders', 'fullSum', 'DiscountSum']].copy()
pivot.columns = ['name', 'year', 'month', 'orders', 'sales', 'discount']
pivot['sales'] = pivot.sales - pivot.discount
```
